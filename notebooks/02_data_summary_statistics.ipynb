# Pheme Dataset - Data Summary Statistics

This notebook provides comprehensive summary statistics for the Pheme rumor detection dataset.

```python
import pandas as pd
import numpy as np

# Display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("Libraries imported successfully!")
```

```python
# Load the dataset
file_path = "../data/processed/pheme_clean.csv"
print(f"Loading dataset from: {file_path}")

try:
    df = pd.read_csv(file_path)
    print(f"Dataset loaded successfully!")
    print(f"Shape: {df.shape}")
    print(f"Columns: {list(df.columns)}")
except Exception as e:
    print(f"Error loading dataset: {e}")
    print("Please ensure the file exists at the specified path.")
    raise
```

## DATA SUMMARY STATISTICS

Key statistics about the dataset structure and composition.

```python
# DATA SUMMARY STATISTICS
print("\n" + "="*60)
print("PHRASE DATASET - COMPREHENSIVE SUMMARY STATISTICS")
print("="*60)

# 1. Total number of events
total_events = df["event_id"].nunique()
print(f"Total events: {total_events}")

# 2. Total number of threads
total_threads = df["thread_id"].nunique()
print(f"Total threads: {total_threads}")

# 3. Total number of source tweets (reply_to is null)
source_tweets = df["reply_to"].isnull().sum()
print(f"Total source tweets: {source_tweets}")

# 4. Total number of reply tweets (reply_to is not null)
reply_tweets = df["reply_to"].notnull().sum()
print(f"Total reply tweets: {reply_tweets}")

# 5. Reply ratio
total_tweets = len(df)
reply_ratio = (reply_tweets / total_tweets) * 100
print(f"Reply ratio: {reply_ratio:.2f}%")

# 6. Label distribution (rumor vs non-rumor)
print("\nLabel distribution (rumor vs non-rumor):")
label_distribution = df["label"].value_counts()
print(label_distribution)

# 7. Show 10 sample rows
print("\n10 sample rows:")
print(df.head(10))

# 8. Additional useful statistics
print("\n" + "="*40)
print("ADDITIONAL STATISTICS")
print("="*40)

print(f"Total tweets: {total_tweets}")
print(f"Average tweets per event: {total_tweets / total_events:.1f}")
print(f"Average tweets per thread: {total_tweets / total_threads:.1f}")
print(f"Source tweet ratio: {(source_tweets / total_tweets) * 100:.2f}%")

# 9. Event details
print("\n" + "="*40)
print("EVENT DETAILS")
print("="*40)
event_details = df.groupby('event_id').agg({
    'post_id': 'count',
    'thread_id': 'nunique',
    'label': lambda x: x.value_counts().to_dict()
}).rename(columns={'post_id': 'tweet_count', 'thread_id': 'thread_count'})
print(event_details)

# 10. Thread details
print("\n" + "="*40)
print("THREAD DETAILS")
print("="*40)
thread_details = df.groupby('thread_id').agg({
    'post_id': 'count',
    'label': lambda x: x.value_counts().to_dict()
}).rename(columns={'post_id': 'tweet_count'})
print(f"Average thread size: {thread_details['tweet_count'].mean():.2f}")
print(f"Median thread size: {thread_details['tweet_count'].median():.2f}")
print(f"Min thread size: {thread_details['tweet_count'].min()}")
print(f"Max thread size: {thread_details['tweet_count'].max()}")

# 11. Label distribution by event
print("\n" + "="*40)
print("LABEL DISTRIBUTION BY EVENT")
print("="*40)
label_by_event = df.groupby('event_id')['label'].value_counts().unstack(fill_value=0)
print(label_by_event)

# 12. Summary table
print("\n" + "="*60)
print("FINAL SUMMARY TABLE")
print("="*60)
summary_data = {
    'Metric': [
        'Total tweets',
        'Total events',
        'Total threads',
        'Source tweets',
        'Reply tweets',
        'Reply ratio (%)',
        'Average tweets per event',
        'Average tweets per thread'
    ],
    'Value': [
        total_tweets,
        total_events,
        total_threads,
        source_tweets,
        reply_tweets,
        f"{reply_ratio:.2f}",
        f"{total_tweets / total_events:.1f}",
        f"{total_tweets / total_threads:.1f}"
    ]
}
summary_df = pd.DataFrame(summary_data)
print(summary_df.to_string(index=False))

print("\n" + "="*60)
print("LABEL BREAKDOWN")
print("="*60)
for label, count in label_distribution.items():
    percentage = (count / total_tweets) * 100
    print(f"{label}: {count} tweets ({percentage:.2f}%)")
```
